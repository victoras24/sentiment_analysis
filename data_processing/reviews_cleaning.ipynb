```python
data_framee  = pd.read_csv('../data_collection/review_data.csv')
```

```python
data_framee = data_framee.drop(columns=['visit_date']).copy()
```

```python
from datetime import datetime

# Convert the "rating_date" column to the desired format
data_framee["rating_date"] = data_framee["rating_date"].apply(lambda date: datetime.strptime(" ".join(date.split(" ")[-2:]), "%d, %Y").strftime("%d/%m/%Y"))

# Print the resulting DataFrame
print(data_framee)
```

```python
fix_issue_date = pd.read_csv('../data_collection/review_data.csv')
fix_issue_date
```

```python
from datetime import datetime

# Convert the "rating_date" column to the desired format
fix_issue_date["rating_date"] = fix_issue_date["rating_date"].apply(lambda date: datetime.\
                                                                    strptime(date.replace("Reviewed ", ""), "%B %d, %Y").\
                                                                        strftime("%d/%m/%Y"))

# Print the resulting DataFrame
print(fix_issue_date)
```
