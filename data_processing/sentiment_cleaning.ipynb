```python
text_new = pd.read_csv('../data_collection/review_test.csv').reset_index()
text_new.head()
```

```python
text_new.drop(columns=['Unnamed: 1' ,'Unnamed: 2'], inplace=True)
text_new.head()
```

```python
data_frame = pd.merge(df_dates_merge, text_new, right_index=True, left_index=True)
data_frame
```

```python
data_frame.drop(columns=['index'], inplace=True)
data_frame.head()
```

```python
example = data_frame['text'][24]
example
```

```python
import pandas as pd
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification

# Load the sentiment analysis pipeline and model
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
sentiment_pipe = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

sentiment_pipe(example)
```

```python
sentiment_pipe('It was disgusting')
```

```python
data_frame['id'] = range(len(data_frame))
data_frame
```

```python
result = {}
max_length = 512 # set the maximum length limit for the transformer model

for i, row in tqdm(data_frame.iterrows(), total=len(data_frame)):
    text = row['text']
    myid = row['id']
    # truncate the longer sentences to fit within the maximum length limit
    input_ids = np.array(tokenizer.encode(text, max_length=max_length, add_special_tokens=True))
    input_text = tokenizer.decode(input_ids, skip_special_tokens=True)
    result[myid] = sentiment_pipe(input_text)
```

```python
sentiment = pd.DataFrame(result).T
sentiment.head()
```

```python
sentiment.reset_index(inplace=True)
sentiment.rename(columns={'index' : 'id'}, inplace=True)
```

```python
sentiment.rename(columns={'index' : 'id'}, inplace=True)
```

```python
sentiment_df = pd.merge(data_frame, sentiment,  how='left')
```

```python
sentiment_df
```

```python
sentiment_df.drop(columns=['id'],inplace=True)
```

```python
sentiment_df = sentiment_df.rename(columns={sentiment_df.columns[6]: 'sentiment'})
sentiment_df
```

```python
sentiment_df['sentiment'][0:10]
```

```python
sentiment_df = sentiment_df.assign(label=sentiment_df.iloc[:, -1].apply(lambda x: x['label']), score=sentiment_df.iloc[:, -1].apply(lambda x: x['score']))
sentiment_df
```

```python
sentiment_df.drop(columns=['sentiment'], inplace=True)
sentiment_df.head()
```

```python
sentiment_df.reset_index(inplace=True)
```

```python
sentiment_df.to_csv('../sentiment_analysis/sentiment_data.csv')
```
