```python
from bs4 import BeautifulSoup

# Define a function to extract the star rating from a class name
def get_star_rating(class_name):
    if "bubble_50" in class_name:
        return 5
    elif "bubble_40" in class_name:
        return 4
    elif "bubble_30" in class_name:
        return 3
    elif "bubble_20" in class_name:
        return 2
    elif "bubble_10" in class_name:
        return 1
    else:
        return None

profile_name = []
user_total_reviews = []
rating_date = []
visit_date = []
stars = []
Summary = []
Text = []

# Loop over all HTML files
for file_number in range(0, 2730, 15):

    with open(f"../data_collection/html_collected/review{file_number}.html", "r") as reader:
        html_string = reader.read()
        soup = BeautifulSoup(html_string, "html.parser")

    for box in soup.find_all(class_="review-container"):
        try:
            profile_name.append(box.find_all(class_="info_text pointer_cursor")[0].text.strip())
        except IndexError:
            profile_name.append(np.nan)
        user_total_reviews.append(box.find_all(class_="reviewerBadge badge")[0].text.strip())
        rating_date.append(box.find_all(class_="ratingDate")[0].text.strip())
        visit_date.append(box.find_all(class_="prw_rup prw_reviews_stay_date_hsx")[0].text.strip())
        Summary.append(box.find_all(class_="quote")[0].text.strip())
        Text.append(box.find_all(class_="prw_rup prw_reviews_text_summary_hsx")[0].text.strip())
        class_name = box.find_all(class_="ui_bubble_rating")[0]["class"][1]
        star_rating = get_star_rating(class_name)
        stars.append(star_rating)

    for box in soup.find_all(class_="review hsx_review"):
        try:
            profile_name.append(box.find_all(class_="info_text pointer_cursor")[0].text.strip())
        except IndexError:
            profile_name.append(np.nan)
        user_total_reviews.append(box.find_all(class_="reviewerBadge badge")[0].text.strip())
        rating_date.append(box.find_all(class_="ratingDate")[0].text.strip())
        visit_date.append(box.find_all(class_="prw_rup prw_reviews_stay_date_hsx")[0].text.strip())
        Summary.append(box.find_all(class_="quote")[0].text.strip())
        Text.append(box.find_all(class_="prw_rup prw_reviews_text_summary_hsx")[0].text.strip())
        class_name = box.find_all(class_="ui_bubble_rating")[0]["class"][1]
        star_rating = get_star_rating(class_name)
        stars.append(star_rating)
```

```python
data = {'profile_name' : profile_name,
        'user_reviews': user_total_reviews,
        'rating_date': rating_date,
        'visit_date': visit_date,
        'summary': Summary,
        'text': Text,
        'stars': stars}
df = pd.DataFrame(data)

df
```
